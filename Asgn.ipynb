```python
import bs4 as b
import urllib,urllib.request as u
```

```python
hisurl = 'http://www.history.com/this-day-in-history'
hispage = u.urlopen(hisurl)
soup = b.BeautifulSoup(hispage,'lxml')
```

```python
soup
```

```python
import datetime as d
print(soup.title.text,d.datetime.now())
```

```python
t=1
for c in soup.findAll('div',{'class':'category'}):
    print(t,' ',c.text)
    t+=1
```

```python
i=1
for title in soup.find('div',{'class':'day-articles'}).findAll('p'):
    print(i)
    i+=1
    print(title.text)
```

```python
def souping(url):
    soup = b.BeautifulSoup(urllib.request.urlopen(url),'lxml')
    return soup
```
